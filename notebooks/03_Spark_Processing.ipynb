## 1. Setup

Import pipeline Spark từ `src/spark_processing.py`.


```python
from pathlib import Path
import sys

import pandas as pd

try:
    PROJECT_ROOT = Path(__file__).resolve().parents[1]
except NameError:
    cwd = Path.cwd().resolve()
    PROJECT_ROOT = cwd if (cwd / "data").exists() else cwd.parent

SRC_DIR = PROJECT_ROOT / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

from spark_processing import run_spark_processing

RAW_CSV = PROJECT_ROOT / "data" / "raw" / "fraudTrain.csv"
TABLE_DIR = PROJECT_ROOT / "reports" / "tables"

print("Project root:", PROJECT_ROOT)
print("Raw CSV:", RAW_CSV)
print("Table dir:", TABLE_DIR)
```

## 2. Run Spark Pipeline

Pipeline thực hiện các bảng:

- Tổng quan dataset.
- Fraud theo `category`.
- Fraud theo `state`.
- Merchant có tỷ lệ fraud cao.
- Fraud theo giờ giao dịch.
- Thống kê amount theo target.
- So sánh Pandas và Spark cho aggregation theo category.


```python
outputs = run_spark_processing(RAW_CSV, TABLE_DIR)
outputs
```
